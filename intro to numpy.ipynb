```python
"""Joining Numpy Arrays
1.vstack().........vertically stack
2.hstack()...........horizondally stack/head stack
3.column_stack()"""
```

```python
#vstack()
```

```python
import numpy as np
n1=np.array([10,20,30])
n2=np.array([20,90,60])
n3=np.array([30,75,87])
np.vstack((n1,n2,n3))
```

```python
#hstack()
```

```python
import numpy as np
n1=np.array([10,20,30])
n2=np.array([20,90,60])
np.hstack((n1,n2))
```

```python
#column_stack()
```

```python
import numpy as np
n1=np.array([10,20,30])
n2=np.array([20,90,60])
np.column_stack((n1,n2))
```

```python
# numpy intersection ...to find the common elements in arrays
```

```python
import numpy as np
n1=np.array([10,45,62])
n2=np.array([10,65,78])
np.intersect1d(n1,n2)
```

```python
# numpy difference.....to find elements present in one particular array not in another
```

```python
import numpy as np
n1=np.array([10,15,20,25])
n2=np.array([15,20])
np.setdiff1d(n1,n2)
```

```python
# ADDITION OF NUMPY ARRAYS
```

```python
# case 1
```

```python
import numpy as np
n1=np.array([10,15,60])
n2=np.array([5,20,12])
np.sum([n1,n2])        #...10+15+60+5+20+12 = 122
```

```python
# case 2 ......vertical addition
```

```python
import numpy as np
n1=np.array([25,15,10])
n2=np.array([5,25,18])
np.sum([n1,n2],axis = 0) #....25+5 ,15+25, 10+18
```

```python
#case 3......horizondal addition
```

```python
import numpy as np
np.sum([[0, 1], [0, 5]], axis=1) #0+0, 1+5
```
